```python
# 导入库
import pandas as pd
from sklearn.linear_model import LogisticRegression

# 加载训练集
train_data = pd.read_csv('/kaggle/input/titanic/train.csv')

# 处理缺失值
train_data['Age'] = train_data['Age'].fillna(train_data['Age'].mean())
for each in train_data.columns:
    train_data[each] = train_data[each].fillna(train_data[each].mode()[0])

# 转换分类变量
train_data['Sex'] = train_data['Sex'].map({'male': 0, 'female': 1})
train_data['Embarked'] = train_data['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})

# 选择特征和标签
X = train_data[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']]
y = train_data['Survived']

# 训练模型
model = LogisticRegression()
model.fit(X, y)

# 评估模型
score = model.score(X, y)
print('Accuracy:', score)
```

```python
# 加载测试集
test_data = pd.read_csv('/kaggle/input/titanic/test.csv')

# 处理缺失值
test_data['Age'] = test_data['Age'].fillna(test_data['Age'].mean())
for each in test_data.columns:
    test_data[each] = test_data[each].fillna(test_data[each].mode()[0])

# 转换分类变量
test_data['Sex'] = test_data['Sex'].map({'male': 0, 'female': 1})
test_data['Embarked'] = test_data['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})

# 对测试集进行预测
X_test = test_data[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']]
y_pred = model.predict(X_test)

# 将预测结果和乘客ID合并成一个新的DataFrame
output = pd.DataFrame({'PassengerId': test_data['PassengerId'], 'Survived': y_pred})

# 保存为csv文件
output.to_csv('submission.csv', index=False)
```
